Oooookay.  What do we want to do with this?  The first thing that comes to mind is to see how many words are spoken per character - but first let's create that function to pull transcript content from a wiki page.  To do this we'll start by simply copy-pasting what we've done so far in the cells above.

```python
# given a url:
def get_transcript(url):
    r = requests.get(url)
    content = r.content
    soup = bs(content, 'html.parser')
    transcript = soup.find('div', {'class':'poem'})
    return transcript
```

We'll also save this function in a .py file so we can import it in another session.  We'll save the transcript URLs for importing as well so we don't have to read a txt file.
