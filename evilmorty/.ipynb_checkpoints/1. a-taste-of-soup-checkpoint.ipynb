# A taste of soup

Given a base URL, locate the URLs for all the transcripts  

---

```python
base = 'http://rickandmorty.wikia.com/wiki/Category:Transcripts'
```

```python
r = requests.get(base)
```

```python
content = r.content.decode('utf-8')
```

```python
soup = bs(content, 'html.parser')
```

`soup.findAll??`

     """Extracts a list of Tag objects that match the given
        criteria.  You can specify the name of the Tag and any
        attributes you want the Tag to have.

        The value of a key-value pair in the 'attrs' map can be a
        string, a list of strings, a regular expression object, or a
        callable that takes a string and returns whether or not the
        string matches for some custom definition of 'matches'. The
        same is true of the tag name."""


```python
[tag.attrs for tag in soup.findAll('a')]
```

```python
transcripts = set([base.split('/wiki/')[0] + tag.attrs['href']
 for tag in soup.findAll('a') 
 if ('href' in tag.attrs.keys()) 
 and ('/Transcript' in tag.attrs['href'])])
```

```python
len(transcripts)
```

```python
transcripts
```

```python
with open('transcript-urls.txt','w') as f:
    for t in transcripts:
        f.write(t+'\n')
```
